```python
def data_info(data):
   cols = data.columns
   unique_values = [data[col].value_counts().head(10).index.to_numpy() for col in cols]
   n_uniques = [data[col].nunique() for col in cols]
   dtypes = [data[col].dtype for col in cols]
   nulss = [data[col].isnull().sum() for col in cols]
   dup = [data.duplicated().sum() for col in cols]
   return pd.DataFrame({"Column": cols, "Data Type": dtypes, "Unique Values": unique_values, "Number of Unique Values": n_uniques, "NaN Count": nulss, "Duplicated Count": dup})
```

```python
def categorical_plots(data, columns, no_of_cat_values):
    total_cols = 3
    total_rows = math.ceil(len(columns) / total_cols)
    # Change the figsize argument to a tuple with two elements
    plt.figure(figsize=(27 , 5 * total_rows), dpi=100)
    plot_index = 1
    for column in columns:
        if len(data[column].unique()) < no_of_cat_values:
            plt.subplot(total_rows, total_cols, plot_index)
            sns.countplot(x=data[column], palette = "Spectral" , hue=data[column])
            plt.title(f"count plot of {column}.")
            plot_index += 1
        else:
            print(f"column {column} is not categorical.")
    plt.tight_layout()
    plt.show()

categorical_plots(df, df.columns, 10)
```

# Skewness

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
def skewness_log_square(data , numbers_of_cat_value):
  features = data.select_dtypes(np.number).columns
  for feature in features:
      if data[feature].nunique() <= numbers_of_cat_value:
          print(f"Categorical featuers {feature}")
          continue

      skewness = data[feature].skew()
      if skewness  > 0.5 :
        transformation = np.log1p
        transformation_name = "log"
      elif skewness < -0.5 :
        transformation = np.square
        transformation_name = "square"
      else :
        print(f"Normal distibution {feature}")
        continue

      logTr = ColumnTransformer(transformers = [(transformation_name.lower(), FunctionTransformer(transformation),[feature])])
      plt.figure(figsize=(15,6))
      plt.subplot(1,2,1)
      plt.title(f"distribution of {feature} before transformation" , fontsize = 15 )
      sns.histplot(data[feature], kde = True , color = "green")
      plt.subplot(1,2,2)

      df_transformed = pd.DataFrame(logTr.fit_transform(data))
      plt.title(f"Distribution of {feature} after transformation" , fontsize = 15)
      sns.histplot(df_transformed, bins = 20 , kde = True ,legend = False)
      plt.xlabel(feature)
      plt.show()

      print(f"Skewness was {round(data[feature].skew(), 5)} before & is {round(df_transformed[0].skew())} after log transformation. ")
      data[feature] = df_transformed[0]
  return data
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
# Step 1: Ordinal Encoding for 'Favorite Color' column
df['Favorite Color'] = df['Favorite Color'].astype('object')
df['Favorite Music Genre'] = df['Favorite Music Genre'].astype('object')
df['Favorite Beverage'] = df['Favorite Beverage'].astype('object')
df['Favorite Soft Drink'] = df['Favorite Soft Drink'].astype('object')
# Ensure 'Favorite Soft Drink' column is treated as object (categorical)

# Step 2:
# order every element in this column by:(from o to infinity) , whatever the the number of the elements in this column
ordinal_encoder = OrdinalEncoder()
df['Favorite Color_encoded'] = ordinal_encoder.fit_transform(df[['Favorite Color']])
df['Favorite Music Genre_encoded'] = ordinal_encoder.fit_transform(df[['Favorite Music Genre']])
df['Favorite Beverage_encoded'] = ordinal_encoder.fit_transform(df[['Favorite Beverage']])
df['Favorite Soft Drink_encoded'] = ordinal_encoder.fit_transform(df[['Favorite Soft Drink']])

# Encode 'Gender' column as binary (1 for 'F', 0 for 'M')
df['Gender_encoded'] = (df['Gender'] == 'F').astype(int)
```
